## I.Introduction:

### DataSet & Description: GLOBAL TERRORISM DATABASE(1970-2016)

The Global Terrorism Database, a systematic Database compiled by National Consortium for the Study of Terrorism and Responses to Terrorism (START) and the Univeristy of Maryland. 

Charecteristics of the GTD DataBase (Start.umd.edu, 2018):

> - Open-source database, information on over 170,000 terrorist attacks.Over 4,000,000 news articles and 25,000 news sources were reviewed to collect incident data from 1998 to 2016 alone 

> - Data on domestic, and international terrorist incidents, occured in the time period of 1970-2016. It is currently the most comprehensive unclassified database on terrorist events in the world. 

> - Reports from different open media sources.The National Consortium for the Study of Terrorism and Responses to Terrorism (START) makes the GTD available, to increase understanding of terrorist violence so that it can be more readily studied and defeated. 

According to the  [Codebook: Inclusion Criteria And Variables](https://www.start.umd.edu/gtd/downloads/Codebook.pdf), the GTD Database has three inclusion categories:



| Codebook: Inclusion Criteria Description  |
|:-------------:|
| CRITERIA 1: The violent act must be aimed at attaining a political, economic, religious, or social goal. This criterion is not satisfied in those cases where the perpetrator(s) acted out of a pure profit motive or from an idiosyncratic personal motive unconnected with broader societal change | 
| CRITERIA 2: To satisfy this criterion there must be evidence of an intention to coerce, intimidate, or convey some other message to a larger audience (or audiences) than the immediate victims. Such evidence can include (but is not limited to) the following: pre‐ or post‐attack statements by the perpetrator(s), past behavior by the perpetrators, or the particular nature of the target/victim, weapon, or attack type.      | 
| CRITERIA 3: The action is outside the context of legitimate warfare activities, insofar as it targets non‐combatants (i.e. the act must be outside the parameters permitted by international humanitarian law as reflected in the Additional Protocol to the Geneva Conventions of 12 August 1949 and elsewhere).      |

## II.Wikipedia Analysis: What is Terrorism? (Tag Cloud)


To understand the general notion towards Terrorism, we build a Tagcloud based on the Term: Terrorism at [Wikipedia Webiste](https://en.wikipedia.org/wiki/Terrorism).

```python
# Step: Import Library:

import sys
from os import path
import numpy as np
from PIL import Image
import wikipedia
from wordcloud import WordCloud, STOPWORDS

#Step2: To Define a path in the current Directory:
direc = path.dirname("__file__")

#Step3: Define a function (get_wiki):

def get_wiki(query):
    ''' For a given query, return the most suitable match'''
    head = wikipedia.search(query)[0]
    page = wikipedia.page(head)
    return page.content

#Step4: Define a function (create_wordcloud):
def create_wordcloud(text):
    '''Returns a Array for the Wordcloud Image'''
    mask = np.array(Image.open(path.join(direc, "cloud.png")))
    stopwords = set(STOPWORDS)
    wc = WordCloud(background_color="black",
             max_words=100, 
             mask=mask,
            stopwords=stopwords)
    wc.generate(text)
    wc.to_file(path.join(direc, "wc.png"))


if __name__ == "__main__":

    text = get_wiki("Terrorism")
    create_wordcloud(text)
    
create_wordcloud(get_wiki("Terrorism"))

#Wordcloud created and stored in the local memory as "wc.png"
from IPython.display import Image
Image(filename='wc.png')
```

*In the Wikipedia Tag cloud, the most prominent words are "Government","Violence", "Political", "Religious," & "Organisation".
We infer that Political, Governmental and Religious Organisation, influences Terrorism around the world.*

*This analysis is inspired from the article of Andreas Mueller on the Data Visualisation, using Wikipedia Text: https://github.com/amueller/word_cloud/blob/master/wordcloud/wordcloud.py*
